```python
from numpy import loadtxt, shape
from matplotlib.pyplot import plot as plt
```

```python
data = loadtxt('../data/falling.txt', comments="//")
```

```python
plt(data[:,0], data[:,1], 'o')
```

se calculará la derivada numérica es decir la velocidad, partiendo de los datos que nos da la posición respecto al tiempo de un filtro de café que cae libremente e incluye la fricción con el aire

```python

from pylab import plot, legend

vel1=[]
vel2=[]
vel3=[]

time1=[]
time2=[]
time3=[]

for t in range (1,27):
    
    v1=(data[t,1]-data[t-1,1])/(data[t,0]-data[t-1,0]) # hacia atras
    vel1.append(v1)
    time1.append(data[t,0])

    
for t in range (1,26):
    
    v2=(data[t+1,1]-data[t-1,1])/(2*(data[t,0]-data[t-1,0]))# centrada
    vel2.append(v2)
    time2.append(data[t,0])
    
for t in range (1,26):
    
    v3=(data[t+1,1]-data[t,1])/(data[t+1,0]-data[t,0])# hacia adelante
    vel3.append(v3)
    time3.append(data[t,0])
    
    
plot (time1,vel1, label='hacia atras')
plot (time2,vel2,label='centrada')
plot (time3,vel3,label='hacia adelante')
legend(loc='lower left')
```

Ahora calcularemos la aceleracion partiendo de los datos obtenidos para las velocidades.

```python

ace=[]
timea=[]
    
for t in range (1,25):
    
    a=(vel1[t]-vel1[t-1])/(data[t,0]-data[t-1,0])
    ace.append(a)
    timea.append (data[t,0])
    
plot (timea,ace,label='aceleracón')
legend(loc='lower left')



    
```
