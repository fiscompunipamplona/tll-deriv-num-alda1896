Aceleración analítica
\begin{equation}
a(t)=-ge^{\frac{c}{m}t}
\end{equation}

velocidad analítica
\begin{equation}
v(t)=\frac{mg}{c}(1-e^{\frac{ct}{m}})
\end{equation}

Ecuación diferencial
\begin{equation}
\frac{\mathrm{d} v}{\mathrm{d} t}=g-\frac{c}{m}v(t)
\end{equation}

```python
#comparaacion entra la solución numerica (usando el método de Euler) 
#y la solucion analitica de la ecuación diferencial 

from pylab import plot, show, legend
from math import exp, fabs

vana=[]
vnum=[]
vnum2=[]
vnum3=[]

time=[]
tim2=[]
tim3=[]

dt1=1
dt2=1e-1
dt3=1e-3
g=9.8
m=68.1
c=12.5
v1=0
v2=0
v3=0

for t in range (21):
    
    v1=v1+((c/m)*v1-g)*dt1    #velocidad numerica
    v=(m*g/c)*(1-exp(c*t/m)) # velocidad analitica
   
    vana.append (v)
    vnum.append (v1)
    time.append (t)
    
for t in range (201):
    
    v2=v2+((c/m)*v2-g)*dt2    #velocidad numerica
   
    vnum2.append (v2)
    tim2.append (t*dt2)
    
for t in range (20001):
    
    v3=v3+((c/m)*v3-g)*dt3    #velocidad numerica
   
    vnum3.append (v3)
    tim3.append (t*dt3)
    

plot (time,vana, label='analitico')
plot (time,vnum, label='dt=1')
plot (tim2,vnum2, label='dt=1e-1')
plot (tim3,vnum3, label='dt=1e-3')
legend(loc='upper right')
show
```
